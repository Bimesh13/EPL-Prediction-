```python
import pandas as pd
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn import cross_validation
from sklearn.ensemble import GradientBoostingClassifier

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('final_dataset1.csv')

data = data[data.MW > 3]


data.drop(['Unnamed: 0', 'HomeTeam', 'AwayTeam', 'HM1', 'HM2', 'HM3', 'AM1', 'AM2', 'AM3', 'Date', 'MW', 'HTFormPtsStr', 'ATFormPtsStr', 'FTHG', 'FTAG',
           'HTGS', 'ATGS', 'HTGC', 'ATGC','DiffLP','DiffPts','DiffFormPts',
           'HM4','HM5','AM4','AM5','HTLossStreak5','ATLossStreak5','HTWinStreak5','ATWinStreak5',
           'HTWinStreak3','HTLossStreak3','ATWinStreak3','ATLossStreak3'],1, inplace=True)

display(data.head())
```

```python
data.shape[1]
```

```python
n_matches = data.shape[0]

n_features = data.shape[1] - 1


n_homewins = len(data[data.FTR == 'W'])

win_rate = (float(n_homewins) / (n_matches)) * 100

# Print the results
print ("Total number of matches: {}".format(n_matches))
print ("Number of features: {}".format(n_features))
print ("Number of matches won by home team: {}".format(n_homewins))
print ("Win rate of home team: {:.2f}%".format(win_rate))
```

```python
X_all = data.drop(['FTR'],1)
y_all = data['FTR']


from sklearn.preprocessing import scale

cols = [['HTGD','ATGD','HTP','ATP','HomeTeamLP','AwayTeamLP']]
for col in cols:
    X_all[col] = scale(X_all[col])
X_all
```
